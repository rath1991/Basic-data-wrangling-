Lets import the necessary libraries 

```python
import numpy as np 
import scipy as sp
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import sklearn as sk
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
```

Having done that lets call our file . In this case this is a .xls file and lets see how it looks like. We will do so using pandas

```python
df=pd.read_excel('Superstore data .xlsx')
df.head(100) #displays first five rows
```

```python
df.tail() #displays last five rows
```

Check for missing values 

```python
df.isnull().sum()
```

We see no missing values. 

Are there any duplicate values?

```python
df.duplicated().sum()
```

There are no duplicates. Lets calculate the average basket size which is also the number of items sold in single purchase i.e.  average number of items purchased on a single visit to the store . If we compute it per customer then :

```python
df.pivot_table(index=['Customer ID', 'Order Date'], values=['Quantity'], aggfunc=np.mean)
```

Alternatively we can use groupby 

```python
 df.groupby(['Category', 'Sub-Category'])['Quantity'].mean()
```

Average Price/Product ID for each store


```python
df.groupby(['Product ID'])['Sales'].mean()
```

Average Store Sales

```python
d1=df.Sales*df.Quantity
round(d1.mean(),2)
```

Average store margin in each segment.

```python
df['margin']=df.Profit/df.Sales
round(df.groupby('Segment')['margin'].mean(),2)
```

Top 5 Category by Sales


```python
df['total sales']=df.Sales*df.Quantity
sales=df.groupby('Sub-Category')['total sales'].mean()
rank_cat=round(sales.sort_values(ascending=False),2)
print(rank_cat[0:5])
rank_cat[0:5].plot(kind='bar')
```
