Lets get an estimate of number of missing values if any. 

```python
df.isnull().sum()
```

Now that is interesting as it returns the the total sum of missing values in each column. In 1961 we have 3539 missing values. In Y2011 we have 104 missing values. Now we need to either get rid of this missing values or fill them with some value. 
We see that here almost all columns have missing values. So instead of dropping them, we can fill the missing values is 0, with a forwards value or a backward values, mean or median.
Now each has their own set of drawbacks. 
Populating missing values with mean or median , forward or backward value will reduce the variability. 
We will try all methods here.

Lets start with dropping columns that have any missing value

```python
df_d=df.dropna(axis=1, how='any')
df_d
```

Here we see that we are only left with two columns that have no missing values. For a large scale temporal study that is unacceptable. So lets just fill the missing values in order to get the entire dataset

Filling any missing value with 0

```python
df1=df.fillna(0)
df1.head(10)
```

Filling any missing data with forward fill.

```python
df2a=df.fillna(method='ffill')
df2a.head(10)
```

We can also interpolate . 

```python
df_interp=df.interpolate(method='linear')
df_interp.head(10)
```

Lets fill missing values with mean of that column 

```python
df_meanfill=df.fillna(df.mean() ['Y1961':'Y2013'])
df_meanfill.head()
```

```python
#Lets just select columns first
#get column address
print(df.columns.get_loc("Y1961"),df.columns.get_loc("Y2013"))
```


In the last operation we figured out an interesting part-- the item code column which was integer was included in arithmetic operations and thereby gave rise to float values. We need to ask ourselves if we require the element, item codes and latitude and longitude. Otherwise we can drop them.  

```python
df_drop=df.drop(['Area Code', 'Item Code', 'Element Code', 'latitude', 'longitude'], axis=1)
```

```python
#slicing over columns 
df_years=df.iloc[:,10:61]
```

```python
#Groupby country and item
d1=df_drop.groupby(['Area','Item']).mean()
d1
```

Lets have some visualization now. We want to look at the countries with highest beer producers. 

```python
df_beer=df_drop[df_drop['Item']=='Beer']
df_beerm=df_beer.groupby('Area')['Y2011'].mean()
df_sort=df_beerm.sort_values(ascending=False)
df_sort
```

Now lets plot these values . Will pick top 50 countries

```python
fig=plt.figure(figsize=(20, 16))
df_sort[0:50].plot(kind='bar')
plt.xlabel('Country',size=20)
plt.ylabel('Beer quantity',size=20)
```

Beer production in USA over time 

```python
Country=['United States of America','Brazil','China,mainland']
d_usa=df_drop[(df_drop['Area']=='United States of America')&(df_drop['Item']=='Beer')]
d_bra=df_drop[(df_drop['Area']=='Brazil')&(df_drop['Item']=='Beer')]

y_usa=d_usa.iloc[:,5:55]
r_usa=np.array(y_usa)

y_bra=d_bra.iloc[:,5:55]
r_bra=np.array(y_bra)
time=np.arange(1961,2011)
fig=plt.figure(figsize=(20, 16))
plt.plot(time,r_usa.flatten(), label='usa')
plt.plot(time,r_bra.flatten(),label='brazil')
leg=plt.legend(loc='upper left', numpoints = 1)
leg.get_frame().set_edgecolor('b')
plt.xlabel('Time')
plt.show()
```
